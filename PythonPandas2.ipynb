# Merging two dataframes



```python
import numpy as np
import pandas as pd
```

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Rahul"],
               "Age": np.random.randint(20, 25, size=5),
               "Sub1_Score": np.random.randint(40, 100, size=5),
               "Sub2_Score": np.random.randint(40, 100, size=5),
               "Sub3_Score": np.random.randint(40, 100, size=5)
              }

left_df = pd.DataFrame(scores_data)
left_df
```

```python
right_data = {"Student": ["Tom", "Katey", "Mak"],
               "Age": [24, 24, 24],
               "Sub4_Score": np.random.randint(40, 100, size=3),
               "Sub5_Score": np.random.randint(40, 100, size=3),
              }

right_df = pd.DataFrame(right_data)
right_df
```

```python
#inner JOin(Intersection)
merged_df = pd.merge(left_df, right_df)
merged_df
```

```python
#Outer Join(Union)
merged_df = pd.merge(left_df, right_df, how="outer")
merged_df
```

```python
#another way
pd.concat([left_df,right_df],join = "outer")
```

```python
pd.concat([left_df,right_df],join = "inner")#does not work
```

```python
#arithmatic operations on dataframes
left_df + right_df
```

## Applying various fucntions on dataframe

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Tom", "Rahul"],
               "Age": np.random.randint(20, 25, size=6),
               "Sub1_Score": np.random.randint(40, 100, size=6),
               "Sub2_Score": np.random.randint(40, 100, size=6),
               "Sub3_Score": np.random.randint(40, 100, size=6)
              }

scores_df = pd.DataFrame(scores_data)
scores_df
```

```python
mean = np.mean(scores_df['Sub1_Score'])
mean
```

```python
np.maximum(scores_df['Sub1_Score'], scores_df['Sub2_Score'])
```

# Applying lambda functions

```python
fn = lambda x : 0.8*x
scores_df['Sub1_Score'].apply(fn)
```

```python
fn = lambda x: x - mean
##for multiple columns
scores_df[['Sub1_Score', 'Sub2_Score', 'Sub3_Score']].apply(fn)
```

# Applying Regular Python fuctions

```python
def standardise(x):
    return (x - x.mean()) / x.std()
```

```python
std_df = scores_df.iloc[: , 2:].apply(standardise)
std_df 
```

```python
scores_df.iloc[: , 2:]
```

"applymap()" Apply a function to a Dataframe elementwise.

This method applies a function that accepts and returns a scalar
to every element of a DataFrame.

```python
std_df.applymap(lambda x: '%0.3f'%x)
```

```python
std_df['Sub1_Score'].map(lambda x: '%0.3f'%x)

std_df
```

```python
scores_df.columns = scores_df.columns.map(str.lower)
```
