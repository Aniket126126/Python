```python
#Setting axis labels to a plot
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = np.random.randn(1000).cumsum()

ax.plot(data)

ax.set_xlabel("Index")
ax.set_ylabel("Value")

ax.set_title("Cummulative sum of random normal data")
plt.show()
```

```python

#Setting ticks and ticklabels

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = np.random.randn(1000).cumsum()

ax.plot(data)
ax.set_xlabel("Index")
ax.set_ylabel("Value")

ax.set_title("Cummulative sum of random normal data")

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(["Start", "1st_Quarter", "2nd_Quarter", "3rd_Quarter", "Last_Quarter"],
                  rotation=45,
                  fontsize="medium")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = np.random.randn(1000).cumsum()

ax.plot(data)
ax.set_xlabel("Index")
ax.set_ylabel("Value")

ax.set_title("Cummulative sum of random normal data")

ax.set_xticks([0, 250, 500, 750, 1000])

plt.show()
```

```python

#Adding legends to a plot

x1 = np.random.randn(1000).cumsum()
x2 = np.random.randn(1000).cumsum()
x3 = np.random.randn(1000).cumsum()

fig, ax = plt.subplots(1,1)

ax.plot(x1, color='g', label='X1')
ax.plot(x2, color='r', label='X2')
ax.plot(x3, color='k', label='X3')

ax.set_xlabel("X")
ax.set_title("X1, X2, X3 for 4 quarters")

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(["Start", "1st_Quarter", "2nd_Quarter", "3rd_Quarter", "Last_Quarter"],
                  rotation=45,
                  fontsize="medium")


ax.legend(loc='best')
plt.show()

```

# Plotting monthly returns for MSFT stocks

```python
msft_df = pd.read_csv('MSFT.csv', index_col=0, parse_dates=True)
msft_df
```

```python
##resample the daily data to monthly data using resample() and BM means by month
msft_df['Close'].resample('BM').mean()
```

```python
monthly_return = msft_df['Close'].resample('BM').mean().iloc[-12:]
monthly_return
```

```python
fig, ax = plt.subplots(1,1)

monthly_return.plot.barh(ax=ax)


ax.set_yticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May',
                   'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                   'Nov', 'Dec'])
ax.plot()
```

Normal Distribution Graph

```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-4, 4, num = 100)
constant = 1.0 / np.sqrt(2*np.pi)
pdf_normal_distribution = constant * np.exp((-x**2) / 2.0)
fig, ax = plt.subplots(figsize=(10, 5));
ax.plot(x, pdf_normal_distribution);
ax.set_ylim(0);
ax.set_title('Normal Distribution', size = 20);
ax.set_ylabel('Probability Density', size = 20);
```
